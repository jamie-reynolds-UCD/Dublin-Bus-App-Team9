```python
import os
import pandas as pd
import datetime as dt
import random
import json
```

```python
# Get the entire trips database
tripsDF = pd.read_csv('rt_trips_DB_2018.txt', sep=';')
```

```python
lines = list(tripsDF['LINEID'].unique())[65:130]
```

```python
for line in lines:
    dirname = line
    os.mkdir(line)
```

```python
tripIDdict = {}
tripIDdfDict = {}
for line in lines:
    triIDdict[line] = sorted(list(tripsDF[tripsDF['LINEID']==line]['TRIPID'].unique()))
    
    # make the list into a dataframe
    tripIDdfDict[line] = pd.DataFrame(data={'TRIPID':tripIDdict[line]})
```

```python
tripID_df = pd.read_csv('rt_leavetimes_DB_2018.txt', sep=';', usecols=['TRIPID'])
```

```python
colList = [i for i in range(tripID_df.shape[0])]

tripID_df['originalIndex'] = colList
```

```python
tripID_joined_dfDict = {}
for line in lines:
    print(line)
    workingDF = tripIDdfDict[line]
    joinedDF = pd.merge(tripID_df, workingDF, on='TRIPID')
    joinedDF.sort_values(by=['originalIndex'], inplace=True)
    joinedDF.reset_index(inplace=True, drop=True)
    tripID_joined_dfDict[line] = joinedDF
```

```python
lineSkipDict = {}

for line in lines[0:7]:
    print(f'Now doing {line}.')
    line_leavetimesIndex = [x for x in tripID_joined_dfDict[line]['originalIndex']]
    skipTo = line_leavetimesIndex[0]
    
    skipList = [i for i in range(skipTo)]

    for i in range(len(line_leavetimesIndex)):
        # to keep track of how many elements in the fortySixA_leavetimesIndex list have been iterated through
        if i % 10000 == 0:
            print(i)
        # if we're not at the last element of the index list
        if line_leavetimesIndex[i] != line_leavetimesIndex[-1]:
            dif = line_leavetimesIndex[i+1]-line_leavetimesIndex[i]
            if dif > 1:
                for j in range((line_leavetimesIndex[i]+1), (line_leavetimesIndex[i+1])):
                    skipList.append(j)
        # if we are at the last element of the list
        else:
            for j in range((line_leavetimesIndex[i]+1), (tripID_df.shape[0])):
                skipList.append(j)
    
    for i in range(len(skipList)):
        skipList[i] += 1
                
    lineSkipDict[line] = skipList
```
